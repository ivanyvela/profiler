### Adding borehole data
Borehole data can be stored in a text file (.dat) or an excel file (.xlsx).

The .dat file should contain a row for each lithology, should be tab delimited and have the case-sensitive headers:

id | utm_x | utm_y | elev | top_depths | bot_depths | lith_names | colors | lith_descriptions

- id is the borehole name, it can be used to label on the plotted profiles
- utm_x and utm_y are the utm coordinates
- elev is the ground elevation, in most cases this will be ignored by the plotting routine, instead the elevation will be taken from the TEM models 
- top_depths is the top depth of the layer, relative to ground level. That is, the first layer should have a top_depth of 0 m.
- bot_depths is the bottom depth of the layer, again relative to the ground level.
- lith_names is the lithology name
- colors is the color to be used for plotting, standard python colors (https://matplotlib.org/stable/gallery/color/named_colors.html) and hex codes will work.
- the lith_descriptions contain a description of the layer, it can be left empty but the header should be there.

The .xlsx file should have exactly the same headers, but each unique borehole should be placed on a new sheet. See example in data folder.

```python
from scipy.interpolate import griddata
def regrid_xyz(x, y, z, grid_resolution):
    """
    Regrid XYZ data onto a regular grid.

    Arguments:
    x -- array-like, the x-coordinates of the data points
    y -- array-like, the y-coordinates of the data points
    z -- array-like, the values associated with the data points
    grid_resolution -- int, the number of grid points along each axis

    Returns:
    xi -- 2D array, the x-coordinates of the regular grid
    yi -- 2D array, the y-coordinates of the regular grid
    zi -- 2D array, the interpolated values on the regular grid
    """
    # Create a regular grid
    xi = np.linspace(np.min(x), np.max(x), grid_resolution)
    yi = np.linspace(np.min(y), np.max(y), grid_resolution)
    xi, yi = np.meshgrid(xi, yi)

    # Reshape x, y, and z for interpolation
    points = np.column_stack((x, y))
    values = np.array(z)

    # Interpolate the data onto the regular grid
    zi = griddata(points, values, (xi, yi), method='linear')

    for i in range(len(xi)):

        for j in range(len(yi)):

            dist = np.min(((xi[i][j] - x) ** 2 + (yi[i][j] - y) ** 2) ** 0.5)

            if dist > 100:

                zi[i, j] = np.nan

    return xi, yi, zi
```

```python

from import_colors import getAarhusCols, getParulaCols

from matplotlib.colors import ListedColormap, LinearSegmentedColormap, BoundaryNorm, Normalize
def get_mean_rho(l_top, l_bot, topo=True):

    rho_means = []

    rhos = model.ttem_models[0]['rhos']
    depths = model.ttem_models[0]['depths']
    elev = model.ttem_models[0]['elev']
  

    for i in range(rhos.shape[0]):

        if topo:

            zi = elev[i] - depths[i, :]

        else:
            zi = - depths[i, :]

        idx = np.where((zi <= l_top) & (zi >= l_bot))[0]

        if idx[0] == 0:

            rho = np.insert(rhos[i, idx], 0, np.nan)
            total_thickness = zi[idx[0]] - l_bot

        else:

            rho = np.insert(rhos[i, idx], 0, rhos[i, idx[0]-1])
            total_thickness = l_top - l_bot

        z = np.insert(zi[idx], 0, l_top)

        z = np.append(z, l_bot)

        sig_mean = np.nansum(np.abs(np.diff(z)) * 1/rho) / total_thickness

        rho_mean = 1/sig_mean

        rho_means.append(rho_mean)

    return rho_means


# %%
layer1 = get_mean_rho(0, -5, topo=False)
layer2 = get_mean_rho(l_top=-5, l_bot=-15)
layer3 = get_mean_rho(l_top=-25, l_bot=-35)
layer4 = get_mean_rho(l_top=-40, l_bot=-60)

x_shift =  577800
y_shift = 6177890
layers = [layer1, layer2, layer3, layer4]
```
